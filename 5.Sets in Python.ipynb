### 5. Sets in Python

----

### 1. Introduction to Sets

A `set` in Python is a collection data type used to store multiple unique value.

It is based on the mathematical concept of a set.

### Key characteristics

1. Unorderd
2. Unidexed
3. Mutable(can add/remove items)
4. Does not allow duplicate values
5. Elements must be immutable(hashable)


```python
# Mutable :- Unhashable
# Immutable: Hashable
```

```python
s = {10,20,30}
print(s)
print(type(s))
```

Note:
Order is not guranteed.

## 2. Creating Sets

Method 1: Using Curly Braces {}


```python
s = {1,2,3}
print(s)
print(type(s))
```

Method 2: Using sets() Constructor



```python
s = set([1,2,3,2,2,2,3])
print(s)
print(type(s))
```

### Important

Empyt set can not be created using {}


```python
s = {}
print(s)
print(type(s))
```

```python
s = {1}
print(s)
print(type(s))
```

```python
# empty:

x = set()
print(x)
print(type(x))
```

### 3. Duplicate Elements

Sets automatcally remove duplicates

Example:


```python
s = {10,20,20,30,40,40}
print(s)
print(type(s))
```

### 4. Data Types Allowed in Sets

Elements must be immutable(hashable)

Allowed:

* int, 
* float,
* string,
* tuple
* boolean
* frozenset

```python
s = {10,25.5,"Python",("Hello world",15),True,False, frozenset([10,20])}
print(s)
```

Not Allowd:

* list
* dictionary
* set

```python
s = {10,25.5,"Python",("Hello world",15),True,False, frozenset([10,20]),[10]}
print(s)
```

```python
s = {10,25.5,"Python",("Hello world",15),True,False, frozenset([10,20]),{"a":"apple"}}
print(s)
```

```python
s = {10,25.5,"Python",("Hello world",15),True,False, frozenset([10,20]),{1,2}}
print(s)
```

## 10. Set Operations
---
1. union
2. Intersection
3. difference
4. symmetric_difference

1. Union

Combines elemnets of both sets and return unique values.

```python
A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
```

```python
# union() method

C = A.union(B)
print(C)
```

```python
# symbol: |  --- pipe

print(A | B)
```

## 2. Intersection

Common elements



intersection() method


```python
A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
```

```python
A.intersection(B)
```

```python
# Symbol: & (Ampersand)

print(A & B)
```

```python
### difference

# a-b  : a 
# b-a: b

A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
```

```python
C = A.difference(B)  # In set A all value return after subtracted from B
print(C)
```

```python
C1 = B.difference(A) # In set B all value return after subtracter from A
print(C1)
```

```python
## symbol: -

A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
```

```python
print(A-B)
```

```python
print(B-A)
```

### 4.symmetric_difference



```python
A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
```

```python
C = A.symmetric_difference(B)
print(C)
```

---

|Operation|Complexity|

|remove|O(1)|

Better than lists for membership testing

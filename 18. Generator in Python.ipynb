### 18. Generator

a `generator` is a special function that returns an iterator and produces values one at a time using yield insited of return.


📌 Simple Defintion
> A generator generates values one by one, only when needed.

### Why Generators Are Needed

❌ Without Generator(List)

```python
def get_numbers():
    return [1,2,3,4,5]

get_numbers()
```

Note:

All values stored in memory at once.

✅ With Generator

```python
def get_numbers():
    for i in range(10):
        yield i

x= get_numbers()
x
```

### Differentiate Between `yield` and `return`

|return|yield|
|----|----|
|Ends function|Pauses function|
|Returns one value| Return multiple values|
|Memory heavy|Memory efficient|

### Generator with loop

```python
def count_up(n):
    for i in range(1,n+1):
        yield i
```

```python
for num in count_up(5):
    print(num)
```

```python
obj = count_up(5)
print(next(obj))
```

### How Generator works

📌 Steps:
1. Function called -> no exection
2. Returns generator object
3. next() resumes exection
4. stops at yield
5. continues from same points

----
### Generator vs Normal Function
|Feature|Generator|Function|
|----|----|----|
|`Uses`|yield|return|
|`Memory`|Low|High|
|`Exceution`|Lazy|Immediate|
|`Values`|One at a time| All at once|

### Example: Even Numbers using Generator

```python
def even_number(n):
    for i in range(n+1):
        if i%2==0:
            yield i
```

```python
obj = even_number(10)
print(next(obj))
```

## Real-World Example: Reading Large File

```python
def read_file(fileName):
    with open(fileName,'r') as file:
        for line in file:
            yield line.strip()
read_file(r"C:\Users\User\Desktop\NursaryRhymes.txt")
```

```python
data = read_file(r"C:\Users\User\Desktop\NursaryRhymes.txt")
print(next(data))
```

### Generator with send()


```python
def gen():
    value = yield 1
    yield value
```

```python
g = gen()
print(next(g))
```

### 3. Lists in Python


---

* Lists are one of the most used data types in Python.

Syntax:

we can use [] 

```python
x = "Hello world"
x[0]   # accessing address value
```

### 3. List Indexing

* Positive Indexing
* Negative Indexing

```python
# Positive Indexing
mix_data = [10, 10.25, True, False, "Hello world", [10,20], (10,20), {1,1,2,3,4},{'a':'apple','b':10},None]
print(mix_data)

print(mix_data[0])
print(type(mix_data[0]))

print(mix_data[1])
print(type(mix_data[1]))

print(mix_data[2])
print(type(mix_data[2]))

print(mix_data[3])
print(type(mix_data[3]))

print(mix_data[4])
print(type(mix_data[4]))

print(mix_data[5])
print(type(mix_data[5]))

print(mix_data[6])
print(type(mix_data[6]))
print(mix_data[7])
print(type(mix_data[7]))
print(mix_data[8])
print(type(mix_data[8]))
print(mix_data[9])
print(type(mix_data[9]))
```

```python
## Negative Indexing

print(mix_data[-1])
print(mix_data[-2])
print(mix_data[-3])
print(mix_data[-4])
print(mix_data[-5])
print(mix_data[-6])
print(mix_data[-7])
print(mix_data[-8])
print(mix_data[-9])
print(mix_data[-10])
# print(mix_data[-11]) IndexError: list index out of range
```

### 8. List Methods 

-----

|Method|Work|
|----|----|
|append(x)|add item at the end|
|extend(iterable)| add multiple items|
|insert(i,x)|insert element with i position with value x|
|remove(x)|Remove first occurrence value of x|
|pop(n)| Remove item at index n|
|pop()| Remove last item|
|clear()|Remove all elements|
|index(x)| It return index position of x value|
|count(x)| It counts x occurrence value|
|sort()| Sort the list:ascending|
|reverse()|Reverse the list|
|copy()| Return shallow copy of element|

## 9. index() and Count()

```python
data = [10,20,30,40,50,60,70,80,90,90,80,70,90,90]
print(data)

print(data.index(20))  # 1
print(data.index(80))  # 7
print(data.count(20))
print(data.count(80))
print(data.count(90))
```

## 10. Sorting List

#### Ascending Sort


```python
nums = [50,10,40,30]
print("Original List:",nums)

nums.sort()
print(nums)
```

## 11. Reverse List

### reverse value only

```python
nums = [50,10,40,30]
print("Original List:",nums)

nums.reverse()  # method reverse()
print(nums)
```

```python
### Descending 

nums = [50,10,40,30]
print("Original List:",nums)

nums.sort(reverse=True)  # method sort under reverse parameter
print(nums)
```

```python
nums = [50,10,40,30]
print("Original List:",nums)

nums.sort()  # mehtod sorting
print(nums)
```

```python
x = nums[::-1]  # logic of slicing
print(x)
```

### 16. Membership Operator: [in or not in]

```python
nums = [10,20,30]

print(10 in nums)
print(10 not in nums)

print(20 not in nums)
print(20 in nums)
```

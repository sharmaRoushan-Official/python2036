## 2. String in Python

## 1. What is a String in Python?

A `string` in Python is a sequence of Unicode character enclosed in quotes.

```python
s1 = "Hello"  # double quote
s2 = 'Hello'   # single quote
s3 = """Hello"""  # triple quote: Multi-line comment
s4 = '''Hello'''   # triple quote: Multi line comment 
print(s1)
print(s2)
print(s3)
print(s4)
```

#### Key points:

## 7. String Method 
---

🔷 Case Converstion
---

```python
s = "hELLo wORld"  # random case
print("Origial Value:",s)

# upper()
s1 = s.upper()
print("Upper:",s1)


# lower()

s2 = s.lower()
print("Lower:",s2)

# capitalize: first character is in upper 
s3 = s.capitalize()
print("Capitalize:",s3)


# title: first character is in capital and after space also
s3 = s.title()
print("Title:",s3)

# swapcase
s4 = s.swapcase()
print("SwapCase",s4)
```

### Searching and Finding
----


```python
s = "hELLo wORld"  # random case
print("Origial Value:",s)

print(s.find('a'))  # -1 if no value found inside the string

print(s.find('h'))  # 0 is the positiion of h
print(s.find('l'))
print(s.find('L'))
print(s.find('Lo'))

print(s.index('h'))  # same as it is find but a little bit different
print(s.index('l'))
print(s.index('L'))
print(s.index('Lo'))
print(s.index('a'))  # if no value found inside the string it will be return valueError
```

```python
## Checking Methods

# isalpha()
# isdigit()
# isalnum()
# isspace()
# islower()
# isupper()

# ------

s = "hello world"  # alphanetical
s1 = "123"
s2 = "hello123"
s3 = " "
s4 = "HELLO"
print(s.isalpha())
print(s1.isdigit())
print(s2.isalnum())
print(s3.isspace())
print(s.islower())
print(s4.isupper())
```

```python
### Trimming:


# strip()
# lstrip()
# rstrip()

str1 = "            Hello World                   "
print(str1)

str2 = str1.strip()  # front+last remove white spaces
print(str2)

str3 = str1.lstrip()  # left spaces remove
print(str3)

str4 = str1.rstrip()  # right spaces remove
print(str4)
```

```python
### UTF-8  #Unicode Transformation Format

s = "hello" # Normal Text   01101000 01100101 01101100 01101100 01101111
# 01101000 01100101 01101100 01101100 01101111
print(s) 
encoded = s.encode("utf-8")
print(encoded)

decoded = encoded.decode('utf-8')
print(decoded)
```

```python
def welcome_decorator(func):
    def wrapper():
        print("Welcome")
        func()
        print("ThankYou")
    return wrapper
@welcome_decorator
def greet():
    print("Hello,Gayatri!")
greet()
```

```python
import time
def execution_time(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print("Execution Time:", end - start, "seconds")
    return wrapper
@execution_time
def task():
    for i in range(100000):
        pass
task()
```

```python
logged_in = True
def login_required(func):
    def wrapper():
        if logged_in:
            func()
        else:
            print("Please login first.")
    return wrapper
@login_required
def dashboard():
    print("Welcome to Dashboard!")
dashboard()
```

```python
def call_counter(func):
    count = 0
    def wrapper():
        nonlocal count
        count += 1
        print("Function called", count, "times")
        func()
    return wrapper
@call_counter
def hello():
    print("Hello!")
hello()
hello()
hello()
```

```python
def uppercase(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper
@uppercase
def message():
    return "Welcome to Python"
print(message())
```

```python
def repeat(func):
    def wrapper():
        for i in range(3):
            func()
    return wrapper
@repeat
def greet():
    print("Hello!")

greet()
```

```python
def logger(func):
    def wrapper(*args):
        print("Arguments:", args)
        result = func(*args)
        print("Result:", result)
        return result
    return wrapper
@logger
def add(a, b):
    return a + b
add(10, 20)
```

```python
def even_only(func):
    def wrapper(num):
        if num % 2 == 0:
            func(num)
        else:
            print("Only even numbers are allowed.")
    return wrapper
@even_only
def display(num):
    print("Even Number:", num)
display(8)
display(5)
```

```python
def password_protect(func):
    def wrapper():
        password = input("Enter Password: ")
        if password == "admin123":
            func()
        else:
            print("Access Denied.")
    return wrapper
@password_protect
def secret():
    print("Welcome! Access Granted.")
secret()
```

```python
def exception_handler(func):
    def wrapper(*args):
        try:
            return func(*args)
        except Exception as e:
            print("Exception:", e)
    return wrapper
@exception_handler
def divide(a, b):
    return a / b
print(divide(10, 2))
divide(10, 0)
```

```python
def positive_only(func):
    def wrapper(num):
        if num > 0:
            func(num)
        else:
            print("Invalid Number.")
    return wrapper
@positive_only
def display(num):
    print("Number:", num)
display(10)
display(-5)
```

```python
role = "admin"
def authorize(func):
    def wrapper():
        if role == "admin":
            func()
        else:
            print("Unauthorized Access.")
    return wrapper
@authorize
def settings():
    print("Welcome Admin!")

settings()
```

```python
def reverse_string(func):
    def wrapper():
        result = func()
        return result[::-1]
    return wrapper
@reverse_string
def message():
    return "Python"
print(message())
```

```python
def multiply_by_10(func):
    def wrapper():
        result = func()
        return result * 10
    return wrapper
@multiply_by_10
def number():
    return 5
print(number())
```

```python
def minimum_age(func):
    def wrapper(age):
        if age >= 18:
            func(age)
        else:
            print("Access Denied. Age must be 18 or above.")
    return wrapper

@minimum_age
def vote(age):
    print("You are eligible to vote.")

vote(20)
vote(15)
```

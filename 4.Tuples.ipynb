## 4. Tuples in Python

----

A `tuple` in Pyton is built-in data structure used to store multiple values in a single variable.


⭐ Tuple is `Immutable` Data type

---

#### Key Characterstics

--
* Ordered collections
* Immutable(Cannot be changed or modified after creation)
* Allows duplicate values
* Can sotre different data types
* Faster than lists
* Written using round brackets () -->> Parentheses

```python
numbers = (10,20,30)
print(numbers)
print(type(numbers))
```

### 2. Creating Tuples
---

### Mehtod-1 --> Using parentheses


```python
fruits = ('apple',"banana","mango")
print(fruits)
```

### Method-2: Without Parentheses

Parentheses are optional

```python
colors = "read","green","black"
print(colors)
print(type(colors))
```

#### Method-3 - Using tuple() --> Constructor



```python
numbers = tuple([1,2,3,4,5,6])
print(numbers)
print(type(numbers))
```

```python
print(fruits[0])
print(fruits[1])
print(fruits[2])
# print(fruits[3])  # tuple index out of range
```

### 7. Tuple Immutability

Tuple cannot be modified after creation


```python
numbers = (10,20,30)

numbers[0] = 100
```

### 8. Changing Tuple Elements

Since tuple are immutable, convert into list


```python
numbers = (10,20,30)
print(numbers,type(numbers))

temp = list(numbers)
print(temp,type(temp))

temp[0] = 100
print(temp)
```

## 12. Tuple Unpacking

Extract tuple values into variable

```python
data = "john",25,"Engineer"
print(data)
print(type(data))


# Unpacking 
name,age,job = data 
print(name)
print(age)
print(job)
```

```python
data = "john",25,"Engineer",10
print(data)
print(type(data))


# Unpacking 
name,age,job = data 
print(name)
print(age)
print(job)
```

```python
data = "john",25,"Engineer"
print(data)
print(type(data))


# Unpacking 
name,age,job,address = data 
print(name)
print(age)
print(job)
```

### 13. Extended Unpacking

```python
numbers = (1,2,3,4,5)

print(numbers,type(numbers))

a,b,c = numbers
print(a)
print(b)
print(c)
```

```python
# Asterisk (*)
numbers = (1,2,3,4,5)

print(numbers,type(numbers))

a,*b,c = numbers
print(a)
print(b)
print(c)
```

```python
# Asterisk (*)
numbers = (1,2,3,4,5)

print(numbers,type(numbers))

a,b,*c = numbers
print(a)
print(b)
print(c)
```

```python
# Asterisk (*)
numbers = (1,2,3,4,5)

print(numbers,type(numbers))

*a,b,c = numbers
print(a)
print(b)
print(c)
```

## 21. Tuple VS List

---

|Feature|Tuple|List|
|---|----|---|
|Brackets|()|[]|
|Mutable|No|Yes|
|Speed|Faster|Slower|
|Momery|Less|More|

### Example

```python
student = ("Ram",21,"Computer Science",85)
print(student)

name,age,course,marks = student
print("Name:",name)
print("Age:",age)
print("Course:",course)
print("Marks:",marks)
```

### 23. When to use Tuples

```python
def simple_interest(p,r,t):
    si = (p*r*t)/100
    totalAmount = p+si
    print(f"SI value is:{si}\nTotal Actual is:{totalAmount}")
simple_interest(1000,20,1)
```

```python
simple_interest(1000000,20,1)
```

```python
### Write a function convert_celsius_to_fahrenheit(c)

# (0°C × 9/5) + 32 = 32°F

def convert_celsius_to_fahrenheit(c):
    fah = (c*9/5)+32
    return f"Fahrenheit value is:{fah} °F"

convert_celsius_to_fahrenheit(37)
```

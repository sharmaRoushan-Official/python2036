### Variables and Expressions

#### A varibale is name used to sotre data in memory.[RAM- Random Access Memory]

### A variable as labelled container that holds information.



```python
name = "Rahul"
age = 25
salary = 44444.4

print(name)
print(type(name))

print(age)
print(type(age))

print(salary)
print(type(salary))
```

Here:

* `name` stores text value.
* `age` storen an integer value
* `salary` stores a decimal value

### Rules for Creating Variables in Python

1. Variable Name Must start with:

✅ Letters(a-z, A-Z)

✅ Underscore( _ )

```python
name = "Raj"
print(name)

NAME = "Raj"  # wrong 
print(NAME)

_age = 10
print(_age)

x1 = 100
print(x1)
```

```python
# 2. Variable Name cannot start with Number 

1name = "Raj"  # invalid
```

```python
name1 = "Raj"  # Valid
print(name1)
```

```python
## 3. No Spaces allowd

first name = "vinod"
print(first name)
```

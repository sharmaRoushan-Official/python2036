## Data Typecasting

```python
# 1. Integer

# 2. Float


# 3. Boolean
```

```python
# convert integer value into float and string
num1 = 100    # float, string
print(num1,type(num1))

float1 = float(num1)
print(float1)
print(type(float1))

str1 = str(num1)
print(str1)
print(type(str1))
```

```python
# convert value int integer and string

float1 = 235.66
print(float1)
print(type(float1))

int1 = int(float1)
print(int1)
print(type(int1))


str1 = str(float1)
print(str1)
print(type(str1))
```

```python
# Boolean

bool1 = True  # int, float, string
bool2 = False  # int, float, string
print(bool1)
print(type(bool1))


int1 = int(bool1)
print(int1)
print(type(int1))

float1 = float(bool1)
print(float1)
print(type(float1))

str1 = str(bool1)
print(str1)
print(type(str1))
```

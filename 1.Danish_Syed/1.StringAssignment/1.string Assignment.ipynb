```python
str2 = "madam"
list2 = list(str2) #aage ja ke ulta ho jayega
list3 = list(str2) #sida
list2.reverse() #ulta
list4 = list2 #ulta 

correct_str = str(list3)
rev_str = str(list4)
print("Correct string ",correct_str,type(correct_str))
print("Reverse string ",rev_str,type(rev_str))
print(correct_str == rev_str)
```

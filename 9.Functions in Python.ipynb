### Local Variable

```python
def test():
    x = 10  # local variable: Access inside function only: You can't access this variable outside the function
    return x
test()
```

```python
x
```

```python
x = 100

def test():
    print(x)  # printing inside the function and variable is accessed because it is global

test()
```

```python
x
```

global Keyword

```python
def change():
    x1 = -55
    return x1
change()
```

```python
x1
```

```python
def change():
    global x1
    x1 = -55
    return x1
change()
```

```python
x1
```

## 9. Nested Functions

function inside another function

```python
def outerFunc():
    def inner():
        print("Inner Function Calling")
        def inner2Func():
            print("Inner 2 Function is calling")
        inner2Func()
    inner()
outerFunc()
```

```python
# Hint : Bubble sort 


def max_value_List():
    list1 = [10,5,4,8,9,66,77,-20,33,55,88,14,95,97] 
    maxValue = list1[0]
    for i in list1:
        if maxValue<i:
            maxValue = i
        # print(i)
    print(maxValue)

max_value_List()
```

Q. What is For Loop?

A `for loop` is used to iterate over a sequence(such as a list, tuple, string, dictionary,set or range) and execute a block of code repeatedly.

syntax:


for variable in sequence:

    # code block

Flow:

 * Python takes the first item from the sequence.
 * Assigns it the the loop variable.
 * Executes the code block.
 * Moves to next item.
 * Repeates until all items are processed.

```python
fruits = ["apple","Mango","Grapes","Banana","Orange","Strawberry","blueberry"]


for i in fruits:
    print(i)
```

Explanation

|Iteration|Fruit Value|
|----|----|
|1|apple|
|2|mango|
|3|Grapes|
|4|Banana|
|5|Orange|
|6|Strawberry|
|7|blueberry|

syntax:

while condition:
    
    # code block

#### Flow Diagram:

↓

↓

True ? -------No ------> Exit Loop

↓Yes

Execute code

↓

Back to condition

```python
# 24. Remove all negative numbers from a list using a loop 

numbers = [10, -5, 3, -1, 0, 42, -88]
print("Original List items:",numbers)
positiveList = []
negativeList = []
for i in numbers:
    # print(i)
    if i>=0:
        positiveList.append(i)
    else:
        negativeList.append(i)
print("Positive List items:",positiveList)
print("Negative List items:",negativeList)
```

```python
numbers = [10, -5, 3, -1, 0, 42, -88]

positiveList = []
negativeList = []
i = 0

while i<len(numbers):
    if numbers[i]>0:
        positiveList.append(numbers[i])
    # print(numbers[i])
    else:
        negativeList.append(numbers[i])
    i+=1
print("Positive List is:",positiveList)
print("Negative List is:",negativeList)
```

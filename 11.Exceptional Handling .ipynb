### 6.Handling specifc Exceptions

Always prefer handling specific exceptions, not general.

```python
try:
    a = 10
    b = 0
    print(a/b)
except ZeroDivisionError:  # specific error handle only zerodivisionError.
    print("Division by zero is not allowed.")
except TypeError:
    print("Check the data of value.")
```

### 7. Multiple Except Blocks.

You can handle multiple errors.

    

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x/y)
except ValueError:
    print("Invalid Input, Please enter only number")
except ZeroDivisionError :
    print("you cannot divide by zero")
```

## 8. Single Except Handling Multiple Exception

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x/y)
except (ValueError,ZeroDivisionError):
    print("Invalid Input or Division by zero")
```

### 9. Using Exception as Object( as e)

This is used to display actual error message with single except

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x/y)

except Exception as e:
    print("Error Occurred:",e)
```

### 10. try-except-else Block

else block runs only if no exception occurs.

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    res = x/y
except Exception as e:
    print("Error Occurred:",e)
else:
    print("Result is:",res)
```

Flow:

* if exception - except runs
* if no exception - else runs

## 11. try-except-finally Block

finally block always runs(wheter excption happens or not.)

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    res = x/y
except Exception as e:
    print("Error Occurred:",e)
else:
    print("Result is:",res)

finally:
    print("Program Finished")
```

Note:

* if file exists --> file content+ "Program finished"

* if file not exists ---> "File Not found"+"Program finished"

### 13. Nested try-except

When exception handling is needed inside another try.

```python
try:
    # outer exception
    a = int(input("Enter A: "))
    b = int(input("Enter B: "))
    try:
        # inner exception
        print(a/b)
    except Exception as e:
        print("Error Occurred:",e)
except ValueError:
    print("ValueError")
```

## 17. Custom Excetpions (User-Defined Exceptions)

You can create your own excption class


Example

```python
class InvalidAgeError(Exception):
    pass
```

```python
age = int(input("Enter age: "))

if age<18:
    raise InvalidAgeError("Age must be 18 or above")
print("Vote Now")
```

```python
class NiteshError(Exception):
    pass
```

```python
name = input("Enter Name: ")

if name != "Nitesh":
    raise NiteshError(f"name is not Nitesh:You Entered {name}")
print("Welcome Nitesh")
```

### 18. Custom Exception with Constructor



```python
class BankError(Exception):
    def __init__(self,message):  # constructor
        self.message = message

# super().__init__(self.message)
balance = 500

withdraw = int(input("Enter withdraw amount: "))

if withdraw> balance:
    raise BankError("Insufficient Funds!!")
print("Withdraw Successful")
```

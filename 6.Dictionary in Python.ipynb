### 6. Dictionary in Python 
---

A `dictionary` is a colleciton of key-value pairs.

* Each key is unique
* Each key maps to a value.
* Dictionaries are mutlable(unhashable/changeable)
* They are written using curly braces.{}

```python
student = {
    "name":"Nitesh",
    "age":25,
    "course":"Python"
}
print(student)
```

name,age,course ---> Keys

Nitesh,25,Python ----> Values

### 2. Why We Use Dictionries
----
Use dictionaries when data has a relationship between key and value.

Example:

|Key|Value|
|---|---|
|name|Nitesh|
|age|25|
|course|Python|

Example:



```python
employee ={
    "id":101,
    "name":"Amit",
    "salary":50000
}
print(employee)
```

### 3. Characteristics of Dictionary

Important point of remember

----

#### 1. Keys must be unique

```python
data = {
    'name':'Amit',  # old
    "name":"Nitesh"  # new
}
print(data)
```

last value overwrites the previous one

### 2. Key must be immutable

-----
Allowd:

* string
* number
* tuple
* boolean


Not Allowd:

* list
* dictionary
* set

Example:

```python
data ={
    1 : "Python",
    "name":"Nitesh",
    (1,2):"Tuple Key",
    True:"Set Key",
    False:"Boolean key"
}
print(data)
```

### 3. Values can be anything

----
Values can be:

* string
* number
* list
* tuple
* dictionary
* object

Example:

```python
data ={
    "name":"Nitesh",
    "marks":[80,85,11],
    'address':{
        "city":"Pune",
        "state":"Maharashtra"
    }
    
}
print(data)
```

### 11. Looping Through Dictionary

Loop through keys:



```python
student = {
    "name":"Nitesh",
    "age":55,
    "city":"Ashok Nagar"
}
print(student)
```

```python
for key in student:
    print(key)
```

```python
for key in student.keys():
    print(key)
```

```python
# values

for value in student.values():
    print(value)
```

```python
### both key and value

for key,value in student.items():
    print(key,":",value)
```

### 12. Nested Dictionary

Dictionary inside dictionary

```python
students = {
    "student1":{
        "name":"Aarohi",
        'age':20,
    },
    "student2":{
        "name":"Nitesh",
        'age':55
    }
}
print(students)
```

### 13. Dictionary comprehension

Short way to create dictionary.

```python
sqr = {x:x*x for x in range(10)}
print(sqr)
print(type(sqr))
```

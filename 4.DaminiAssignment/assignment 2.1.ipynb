```python
salary = 50000
years = 8
if years > 10:
    bonus = salary * 0.10
elif years >= 6:
    bonus = salary * 0.08
else:
    bonus = salary * 0.05

total_salary = salary + bonus

print("Bonus =", bonus)
print("Total Salary =", total_salary)
```

```python
a = 10
b = 5
op = "+"
if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
elif op == "^":
    print(a ** b)
else:
    print("Invalid Operator")
```

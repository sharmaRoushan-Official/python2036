### 15. Lambda Functions in Python


#### What is a Lambda Functions?

A lambda function is a small, anonymous[without name or nameless] function written in one single line.


📌 Simple Definition

_A lambda fucniton is one-line function without a name._

#### Why Lambda funcitons are Used?

- ✓ Short & clean code
- ✓ Used for temporary operations
- ✓ Common with `map()`, `filter()`, `reduce()`, `apply()`
- ✓ Improves readability in simple logic.

----

## 🔸Normal Function VS Lambda Function

### Normal Function

```python
def square(x):  # Function name is square()
    return x*x
square(10)
```

```python
square(1000)
```

#### Lambda Function : Nameless 

```python
(lambda x:x*x)(10)  # lambda function 
```

```python
(lambda x:x*x)(100) # lambda function
```

Note:

- ✓ Same Output
- ✓ Less Code 

### syntax of Lambda function

(lambda argument: expression)

📌 Rules:

- ✓ Can have multiple arguments
- ✓ Only one expression
- ✓ Automatilly return value
- ✓ No return keyword

### Lambda filter()

Purpose:

Filter elements based on condition

```python
# Normal Function 

numbers = [10,11,12,15,14,16,18,17,19,20,22,24,26,23,25,2,72,82,88,89]

def evenNum():
    for i in numbers:
        if i%2==0:
            print(i,"Even")
        else:
            print(i,"odd")

evenNum()
```

### lambda function

syntax:

filter(function, sequece)


```python
num = [10,11,12,15,14]


def evenNum(x):
    if x%2==0:
        return "Even"

evenNum(11)
```

```python
### lambda with Normal function
num = [10,11,12,15,14]
list(filter(evenNum, num))
```

```python
## with only lambda 
num = [10,11,12,15,14]
tuple(filter(lambda x:x%2==0,num))
```

### Lambda with `map()` 

Purpose:

Apply a function to each element in the a list.


syntax:

map(function, sequence)

```python
num = [10,11,12,15,14]


def sqrNum():
    for i in num:
        print(i*i,end=" ")

sqrNum()
```

```python
def sqrNum1(x):
    sqr = x*x
    return sqr
sqrNum1(10)
```

```python
## lambda with map [Normal Function]
num = [10,11,12,15,14]

list(map(sqrNum1,num))
```

```python
# lambda function
num = [10,11,12,15,14]

list(map(lambda a:a*a,num))
```

### Lambda with reduce()

## Lambda with Sorting



```python
students = [("Ramesh",66,55),("Anita",92,17),("Kiran",78,99),("Ankita",55,108)]
students
```

```python
# Sort by marks 

sorted(students, key= lambda x:x[0])
```

1)Create a tuple containing 5 different values and print it.

```python
tuple = (1, 2.5, "Nikhil", True, None)
print(tuple,type(tuple))
```

--------------
2) Write a program to find the length of a tuple.

```python
tuple = (1, 2.5, "Nikhil", True, None)
print(len(tuple))
```

------------
3) Write a program to access the third element of a tuple.

```python
tuple = (1, 2.5, "Nikhil", True, None)
print(tuple[2])
```

---------------
4) Write a program to check whether an element exists in a tuple.

```python
fruits = ("Apple", "Banana", "Papaya", "Orange")
print("Orange" in fruits)
print("Mango" in fruits)
```

-----------
5) Write a program to count the occurrence of a specific value in a tuple.

```python
tuple = (1,2,3,5,4,2,6,9)
print(tuple.count(2))
```

-----------
6) Write a program to find the index of a given element in a tuple.

```python
tuple = (1,2,3,5,4,6,9)
print(tuple.index(4))
```

--------------------
7) Write a program to concatenate two tuples.

```python
t1 = (1,2,3)
t2 = (4,5,6)
print( t1 + t2 )
```

-----------------
8) Write a program to convert a tuple into a list.

```python
num = (1, 2, 3, 4, 5, 6)
print(num, type(num))
values = list(num)
print(values, type(values))
      
```

9) Write a program to find the maximum and minimum values in a tuple.

```python
num = (1, 2, 3, 4, 5, 6)
print("Maximum Value :",max(num))

print("Minimum Value :",min(num))
```

----------
10) Write a program to iterate through all elements of a tuple and print them.

```python
num = (1, 2, 3, 4, 5)
for i in num:
    print(i)
```

### Mixed Practice Questions

1) Accept a string from the user and store each character in a list.
    

```python
text = input("Enter a name :") 
name = list(text)
print(name, type(name))
```

-------------------------
2) Convert a list of names into a tuple.

```python
name = ["Nikhil", "Kunal", "Manoj"]
print(name,type(name))
text = tuple(name)
print(text, type(text))
```

------------------------------
3) Find common elements between two lists.

```python
l1 = [1, 3, 4, 7]
l2 = [3, 5, 6, 9]
list = []
for i in l1:
    if i in l2:
        list.append(i)
print(list) 
```

4) Convert a tuple into a string.

```python
tuple = (1, 2, 3, 4, 5)
print(tuple, type(tuple))
num = str(tuple)
print(num,type(num))
```

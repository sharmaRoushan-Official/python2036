1) Given
    nums- [12, 5, 18, 7, 25, 10]
    Use map() with lambda to add 10 each element.

```python
nums = [12, 5, 18, 7, 25, 10]

list(map(lambda x: x + 10, nums))
```

3. Given:
    nums=[11,20,35,40,55,60]
    Use filter() with lambda to extract numebers divisible by 5 but not by 2.

```python
nums = [11, 20, 35, 40, 55, 60]

list(filter(lambda x: x % 5 == 0 and x % 2 != 0, nums))
```

1. Create a list of numbers from 1 to 100 that are divisible by 3 and 5

```python
num = print("The numbers is Divisible by 3 and 5: ",[x for x in range(1,101) if x%3 ==0 and x%5 ==0])
print(num)
```

2. Given 
    nums = [4,9,16,25,36]
    Create a list containing the square root of each number.

```python
nums = [4,9,16,25,36]
sq_root = [x ** 0.5 for x in nums]
print(sq_root)
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
[x for row in matrix for x in row]
```

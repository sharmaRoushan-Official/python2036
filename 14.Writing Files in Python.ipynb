## 14. Writing Files in Python

`Writing files` means storing data permanently in a file from python Program.

📌 Used for:
* Saving user data
* Logs
* Reports
* Exporing results
* Backup files

----

### File Writing Modes (Very Important)


|Mode|Meaning|
|----|----|
|`w`|Write(creates new/overwrites file)|
|`a`|Append(adds data at end)|
|`w+`|Write+Read|
|`a+`|Append+Read|
|`x`|Creates new file if exists it throw error|

### Steps to Write Files

1. Open the file
2. Write data
3. Close the file

### 1. Writing Files using `write()` (Basic)

```python
file = open(r"C:\Users\User\Desktop\data.txt","w")  # first step

# second step
file.write("Hello Welcome to writing files using python.\n")
file.write("This is the example of writing files in Python.\n")
file.write("Hello world how many people are living on the earth?\n")

# third step 
file.close()
```

Note:

* w modes overwrites existing content.
* Use \n for new line.

📌 Does not add newline automatically

## 2. Witing Using `writelines()` [Multiple Lines]

```python
lines = ["Python\n","java\n","JavaScript\n","TypeScript\n","LiveScript\n","EcmaScript\n","BPCL\n","BASIC\n","DOS\n"]

file = open(r"C:\Users\User\Desktop\data.txt","a")
file.writelines(lines)

file.close()
```

## 3. Best Practice: writing Uisng `with`



```python
with open(r"C:\Users\User\Desktop\data.txt","w") as file:
    file.write("Using with statement \n")
    file.write("File auto closed.")
```

## 6. Wiriting Numbers to file:

```python
numbers = [10,20,30,40,50]

with open(r"C:\Users\User\Desktop\data.txt","w") as file:
    file.writelines(numbers)
```

### Fixed

```python
numbers = [10,20,30,40,50]

with open(r"C:\Users\User\Desktop\data.txt","w") as file:
    # file.writelines(numbers)
    for num in numbers:
        file.write(str(num)+"\n")
```

## write a table of 19 in the file 

```python
with open(r"C:\Users\User\Desktop\data.txt","w") as file:
    # file.writelines(numbers)
    
    num = 19
    for i in range(1,11):
        # print(f"{num}x{i}={num*i}")
        file.write(str(num)+"x"+str(i)+"="+str(num*i))
        file.write("\n")
    
```

### table of 1-10 using while loop

```python
with open(r"C:\Users\User\Desktop\data.txt","w") as file:
    i = 1
    while i<=10:
        # print(i)
        file.write(f"============Table of {i}============\n")
        a= 1
        while a<=10:
            # print(f"{i}x{a}={i*a}")
            file.write(f"{str(i)}x{str(a)}={str(i*a)}")
            file.write("\n")
            a+=1  
        i+=1
```

## 7. Write+Read('w+' Mode)



```python
with open("sample.txt","w+") as file:  # Read and write
    file.write("Hello world")
    file.seek(0)  # tell()
    print(file.read())
```

### 8. Writing Using Loop (Log file Example)




```python
with open("log.txt","w") as file:
    for i in range(3):
        file.write(f"Log entry {i} \n")
```

## 9. Exception Handling while Writing


```python
try:
    with open("/root/protected.txt","w") as file:
        file.write("Safe Writing")
except IOError:
    print("File Error Occurred")
```

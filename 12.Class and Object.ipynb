### 5. Consturctor ( `__init__` )


A constructor initializes object data automatically when the object is created.

def `__init__`(self):

````python
class Student:
    # constructor
    def __init__(self,name,age):
        self.name = name
        self.age = age

nitesh = Student("Nitesh",25)  # Positional Argument
````

````python
nitesh.name
````

````python
nitesh.age
````

Key Points:

* Automatically Invoked
* Used to initialized instance variable

### 7. Types of Method:

1. Instance Method
    > Uses self keyword
    
    > Access instance variables
2. Class Method
    > Uses @classmethod [In-Built decorator]
    
    > Uses cls 
3. Static Method
    > Uses @staticmethod [In-Built decorator]
    
    > No self or cls 

1. Instance Variable

````python
class Student:
    def __init__(self,name):
        self.name = name ### instance variable

obj = Student("Gayatri")
obj.name
````

2. Class Variables

````python
class College:
    Collegename = "DY Patil College"  # Class variable
    def __init__(self,name):
        self.name = name ### instance variable
    @classmethod        
    def studentDetails(cls):
        print(f"My College Name is: {cls.Collegename}")
s1 = College("Nitesh")
s1.name
````

````python
s1.studentDetails()
````

````python
class Student:
    name = "Parul University" # class Variable

    def __init__(self, name,age):
        self.name = name
        self.age = age

    # class variable
    @classmethod  # In-Build Decorator
    def collegeName(cls):
        print(f"Name of college is: {cls.name}")
    
s2 = Student("Gayatri",35)
s2.name
````

````python
s2.collegeName()
````

### 3. Static Method

Uses @staticmethod

No self or cls

````python
class Student:
    name = "Parul University" # class Variable

    def __init__(self, name,age):
        self.name = name
        self.age = age

    # class variable
    @classmethod  # In-Build Decorator
    def collegeName(cls):
        print(f"Name of college is: {cls.name}")

    @staticmethod
    def info():
        print("My self Nitesh")

s3 = Student("Gayatri",36)
s3.name
````

````python
s3.info()
````

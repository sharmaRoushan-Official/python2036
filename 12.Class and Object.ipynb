````python
class College:
    Collegename = "DY Patil College"  # Class variable
    def __init__(self,name):
        self.name = name ### instance variable
    @classmethod        
    def studentDetails(cls):
        print(f"My College Name is: {cls.Collegename}")
s1 = College("Nitesh")
s1.name
````

## a). Single/Simple Inheritance

In single/Simple Inheritance, a child inherits from only one parent class.

📌 Structure

Parent ➽ Child 


### Why it is used:

* Code reusability
* Child class can use parent properties and methods.
* Easy to understand and maintain.

#### Real-life Example:

* Parent: Vehicle
* Child: Car

A Car is a Vehicle




````python
# simple inhertace
class Vehicle:  # class is keyword and Vehicle is name of class  # parent
    x = "Nitesh"
    def start(self):  # def is keyword and start is the name of method 
        print("Vehicle Started") # print is statemenet
    @classmethod
    def classVariable(cls):
        print(f"Class Variable is: {cls.x}")

class Car(Vehicle):  # Car is the child of Vehicle
    def drive(self):
        print("Car is driving")


# obj1 = Vehicle()
# obj1.start()

# obj2 = Car()
# obj2.drive()

obj1 = Car()
obj1.start()  # parent property
obj1.drive() # child property
obj1.classVariable()
````

Note:

Single Inheritance is the most basic and safest inheritance. It avoids complextity and ambiguity.

## 2. Multiple Inheritance

 * In Multiple Inheritance, a child class inherits from more than one parent class.

📌 Structure

```text
Parent1      Parent2
    \        /
     \      /
      Child
```

## Why it is used

* When one class needs feature from multiple classes.
* Helps combine functionalities

## Real-life example

* Parent1:Camera
* Parent2:Phone
* Child:SmartPhone


SmartPhone is both a Phone calling and Camera to take picture.

````python
class Phone:  # parent
    def call(self):
        print("Calling............")
        
class Camera:  # parent
    def click_photo(self):
        print("Photo Clicked")

class SmartPhone(Phone,Camera):  # Child 
    def Internet(self):
        print("Internet Explorer....")

obj1 = SmartPhone()
obj1.Internet()
obj1.call()
obj1.click_photo()
````

Note:
Python Supports Multiple Inheritance, It also support MRO [Method Resoultion Order] to avoid Confusion.

````python
print(SmartPhone.mro())
````

````python
print(Car.mro())
````

### 3. Multilevel Inheritance:

In Multilevel Inheritance, inheritance happens in levels.

A class inherits from a class, and antother class inherits from child class.



📌 Strcuture

```text
Grandparent➽Parent➽Child➽GrandChild
```

## Why it is Used

* To represent hierarchy in real world.
* Step-by-step inheritance of features.

## Real-life Example

* Grandparent: Animal
* Parent: Mammel
* Child: Dog

````python
class Animal:  # level-1
    def eat(self):
        print('Eating.....')
        
class Mammal(Animal):  # level-2
    def breath(self):
        print("Breating....")
    def walk(self):
        print("Walking.....")
        
class Dog(Mammal):  # level-3
    def bark(self):
        print("Barking....")

obj1 = Dog()
obj1.bark()
obj1.walk()
obj1.eat()


````

Note:

The child class gets access to all previous level methods.

### 4. Hierarchical Inheritance

In Hierarchical Inheritance, Multiple child classes inherit from a single parent class.

```text
          Parent
        /     |   \
  Child1   Child2    Child3     
```
### Why it is used?

* One parent has common Functionality
* Different Child use and extend it different.


### Real-life example

Parent: Employee

Children:

* Manager
* Developer
* Tester

All are employees but have different roles.

````python
class Employee:  # parent
    def work(self):
        print("Employee is Working...")

class Manager(Employee):  # parent
    def manage(self):
        print("Manager is Managing.....")

        
class Developer(Employee):  # parent
    def code(self):
        print("Developer is Coding......")

m = Manager()
m.manage()  # child class  # role
m.work()  # parent class


d = Developer()
d.code()  # role different
d.work()  # parent class property
````

Note:

This is very common in software systems(Banking, College, EMS)

### 5. Hybrid Inheritance

Hybrid Inheritance is combination of two or more inheritance types.

* It may incldue:

* Mutiple+Mutilevel

* Hierarchical+Multiple

📌 Strcuture

```text
        A
      /   \
     B     C
      \   /
        D
```
Here:

* B and C inherit from A (Hierarchical)
* D inherits from B and C (Multiple)


### Why it is used

* When a system requres complex relationships
* Common in large framewroks and entprises project.


Example

````python
class A:
    def showA(self):
        print("Class A")

class B(A):  # B is a child of A
    def showB(self):
        print("Class B")

class C(A):  # C is a child of A
    def showC(self):
        print("Class C")

class D(B,C):
    def showD(self):
        print("Class D")


obj = D()
obj.showA()
obj.showB()
obj.showC()
obj.showD()
````

Notes:

Hybrid Inheritance is powerful but can cause confusion due to:

* method conflicts
* ambiguily

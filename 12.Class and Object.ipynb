## a). Single/Simple Inheritance

In single/Simple Inheritance, a child inherits from only one parent class.

📌 Structure

Parent ➽ Child 


### Why it is used:

* Code reusability
* Child class can use parent properties and methods.
* Easy to understand and maintain.

#### Real-life Example:

* Parent: Vehicle
* Child: Car

A Car is a Vehicle




````python
# simple inhertace
class Vehicle:  # class is keyword and Vehicle is name of class  # parent
    x = "Nitesh"
    def start(self):  # def is keyword and start is the name of method 
        print("Vehicle Started") # print is statemenet
    @classmethod
    def classVariable(cls):
        print(f"Class Variable is: {cls.x}")

class Car(Vehicle):  # Car is the child of Vehicle
    def drive(self):
        print("Car is driving")


# obj1 = Vehicle()
# obj1.start()

# obj2 = Car()
# obj2.drive()

obj1 = Car()
obj1.start()  # parent property
obj1.drive() # child property
obj1.classVariable()
````

Note:

Single Inheritance is the most basic and safest inheritance. It avoids complextity and ambiguity.

## 2. Multiple Inheritance

 * In Multiple Inheritance, a child class inherits from more than one parent class.

📌 Structure

```text
Parent1      Parent2
    \        /
     \      /
      Child
```

## Why it is used

* When one class needs feature from multiple classes.
* Helps combine functionalities

## Real-life example

* Parent1:Camera
* Parent2:Phone
* Child:SmartPhone


SmartPhone is both a Phone calling and Camera to take picture.

````python
class Phone:  # parent
    def call(self):
        print("Calling............")
        
class Camera:  # parent
    def click_photo(self):
        print("Photo Clicked")

class SmartPhone(Phone,Camera):  # Child 
    def Internet(self):
        print("Internet Explorer....")

obj1 = SmartPhone()
obj1.Internet()
obj1.call()
obj1.click_photo()
````

Note:
Python Supports Multiple Inheritance, It also support MRO [Method Resoultion Order] to avoid Confusion.

### 3. Multilevel Inheritance:

In Multilevel Inheritance, inheritance happens in levels.

A class inherits from a class, and antother class inherits from child class.



📌 Strcuture

```text
Grandparent➽Parent➽Child➽GrandChild
```

## Why it is Used

* To represent hierarchy in real world.
* Step-by-step inheritance of features.

## Real-life Example

* Grandparent: Animal
* Parent: Mammel
* Child: Dog

````python
class Animal:  # level-1
    def eat(self):
        print('Eating.....')
        
class Mammal(Animal):  # level-2
    def breath(self):
        print("Breating....")
    def walk(self):
        print("Walking.....")
        
class Dog(Mammal):  # level-3
    def bark(self):
        print("Barking....")

obj1 = Dog()
obj1.bark()
obj1.walk()
obj1.eat()


````

Note:

The child class gets access to all previous level methods.

### 4. Hierarchical Inheritance

In Hierarchical Inheritance, Multiple child classes inherit from a single parent class.

```text
          Parent
        /     |   \
  Child1   Child2    Child3     
```
### Why it is used?

* One parent has common Functionality
* Different Child use and extend it different.


### Real-life example

Parent: Employee

Children:

* Manager
* Developer
* Tester

All are employees but have different roles.

````python
class Employee:  # parent
    def work(self):
        print("Employee is Working...")

class Manager(Employee):  # parent
    def manage(self):
        print("Manager is Managing.....")

        
class Developer(Employee):  # parent
    def code(self):
        print("Developer is Coding......")

m = Manager()
m.manage()  # child class  # role
m.work()  # parent class


d = Developer()
d.code()  # role different
d.work()  # parent class property
````

Note:

This is very common in software systems(Banking, College, EMS)

### 5. Hybrid Inheritance

Hybrid Inheritance is combination of two or more inheritance types.

* It may incldue:

* Mutiple+Mutilevel

* Hierarchical+Multiple

📌 Strcuture

```text
        A
      /   \
     B     C
      \   /
        D
```
Here:

* B and C inherit from A (Hierarchical)
* D inherits from B and C (Multiple)


### Why it is used

* When a system requres complex relationships
* Common in large framewroks and entprises project.


Example

````python
class A:
    def showA(self):
        print("Class A")

class B(A):  # B is a child of A
    def showB(self):
        print("Class B")

class C(A):  # C is a child of A
    def showC(self):
        print("Class C")

class D(B,C):
    def showD(self):
        print("Class D")


obj = D()
obj.showA()
obj.showB()
obj.showC()
obj.showD()
````

Notes:

Hybrid Inheritance is powerful but can cause confusion due to:

* `var --> Public Variable`

### 1. Public Varaible

* Accessible everywhere(inside and outside the class.)

````python
class Student:
    def __init__(self):
        self.name = "Nitesh" # Public

obj = Student()
obj.name  # Accessible outside the class directly: No restriction
````

### 2. Protected Variable(_var)

Used inside class and subclass(Convention only, still accessible outside.)

````python
class Student:
    def __init__(self):
        self._name = "Nitesh" # protected

obj = Student()
obj.name 
````

````python
class Student:
    def __init__(self):
        self._name = "Nitesh" # protected

    def protectedVariable(self):
        return f"{self._name} is your name"

obj = Student()
obj.protectedVariable()
````

````python
class Student:
    def __init__(self):
        self._name = "Nitesh" # protected

    def protectedVariable(self):
        return f"{self._name} is your name"

obj = Student()
obj._name  # Possible but not recommmeded.
````

### 3. Private Variable(__var)

Accessible only inside the class.



````python
class Student:
    def __init__(self):
        self.__marks = 90 


obj = Student()
obj.__marks
````

````python
class Student:
    def __init__(self):
        self.__marks = 90 


    def accessingPrivate(self):
        return f"Your marks is: {self.__marks} "


obj = Student()
obj.accessingPrivate()
````

### Getter and Setter method

````python
class Student:
    def __init__(self):
        self.name = "Nitesh" # public
        self._age = 30  # protected
        self.__marks = 0  # private

    # marks display [Getter]
    def getMarks(self):
        return f"Your Current marks is:{self.__marks}"

    # set marrks [setter]
    def set_marks(self):
        value = int(input("Enter Marks: "))
        if value>=0 and value<=100:
            self.__marks = value 
            print("Your marks is updated Successfully")
        else:
            print("Invalid Marks Value:",value)

        
s1 = Student()
print(s1.name)
print(s1._age)
# print(s1._Student__marks)

# print(s1.set_marks())
print(s1.getMarks())
````

### Encapsulation Using @property Decorator (Advanced Logic)


````python
class Employee:
    def __init__(self):
        self.name = "Nitesh"
        self.__salary = 10

    @property
    def displayName(self):
        return f"Your name is: {self.name}"

    @property
    def salary(self):
        return self.__salary

    def getAmount(self):
        return f"Your Current Salary is: {self.__salary}"

    @salary.setter 
    def updateSalary(self):
        amount=int(input("Enter your current salary: "))
        if amount>0:
            self.__salary = amount
        else:
            print("Amount is not valid")

e1 = Employee()
# e1.salary()
e1.updateSalary()
````

## 3. Polymorphism in OOPs

Polymorphism means "Many Forms"


In Object-Oriented Programming, the same method name and same behaves differently depending on the object that calls it.


### In simple term:

    * One function or method, but multiple behaviors.


### Why Polymorphism is important

* Code become flexible.
* Code become reusable.
* Suppors Real-world scenarios
* Makes program easy to extend

  * Employee

### Same Person, different behavior based on the role that is called polymorphism.

#### Types of Polymorphism

* Method Overriding
* Polymorphism using Inheritance

* Opertor Overloding

### Method Overloading(Python Style)

Python does not support traditional Overloading But we can achieve it using defualt arguments.

````python
# Overloading - name is same but different parameter
class Calculator:
    def add(self,a,b=0,c=0): # Default 
        return a+b+c

calc = Calculator()
print(calc.add(10))  # a = 10,b=0,c=0  # one argument
print(calc.add(10,13))  # a = 10,b=0,c=0  # two argument
print(calc.add(10,13,78))  # a = 10,b=0,c=0  # three argument
print(calc.add(10,13,78,78))  # a = 10,b=0,c=0  # three argument
````

````python
# Overloading - name is same but different parameter
class Calculator:
    def add(self,*args): # Default 
        return sum(args)

calc = Calculator()
print(calc.add(10))  # a = 10,b=0,c=0  # one argument
print(calc.add(10,13))  # a = 10,b=0,c=0  # two argument
print(calc.add(10,13,78))  # a = 10,b=0,c=0  # three argument
print(calc.add(10,13,78,78))  # a = 10,b=0,c=0  # four argument
print(calc.add(10,13,78,78,45))  # a = 10,b=0,c=0  # five argument
````

Note:

* Same method name add
* Different number of arguments
* Behavior changes automatically

### 2. Method Overriding

* Child class provide its own implementation of parent method.

````python
class Employee:
    def role(self):  # method name   parameter is one self: Same signature
        return "I am an Employee."

class Trainer(Employee):
    def role(self):  # method name   parameter is one self : Same signature
        return "I am a Trainer"

    def TrainerName(self):
        return f"Myself Nitesh"

class TeamLeader(Trainer):
    def role(self):   # method name   parameter is one self  : Same signature
        return "I am Leading the Team"


obj = Trainer()
print(obj.role())  # child 
print(obj.TrainerName())  # child 

# obj2 = 

obj = TeamLeader()
obj.role()
````

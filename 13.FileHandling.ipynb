### 6. read(n) Method 

Reads n characters only.

Example:

```python
filepath = "C:\\Users\\User\\Desktop\\data.txt"
print(filepath)  # file path

# Total 3 steps to open file read 
# step:1 open()

file = open(filepath,"r")
# print(file)  

# step:2 read()

content = file.read(10)  # this will return only 10 characters
content1 = file.read(10) # this will return next 10 characters
print(content)
print(content1)

# step: 3 close

file.close()  # closed file.
```

## 7. readlines() Method

Reads one line at a time.

Example

```python
filepath = "C:\\Users\\User\\Desktop\\data.txt"
print(filepath)  # file path

# Total 3 steps to open file read 
# step:1 open()

file = open(filepath,"r")
# print(file)  

# step:2 readline()

content = file.readline()
content1 = file.readline()
content2 = file.readline()
content3 = file.readline()
content4 = file.readline()
content5 = file.readline()
print(content)
print(content1)
print(content2)
print(content3)
print(content4)
print(content5)

# step: 3 close

file.close()  # closed file.
```

### 8. readlines() Method

Reads all lines into a list.

Example

```python
filepath = "C:\\Users\\User\\Desktop\\data.txt"
print(filepath)  # file path

# Total 3 steps to open file read 
# step:1 open()

file = open(filepath,"r")
# print(file)  

# step:2 read()

content = file.readlines()
print(content)

# step: 3 close

file.close()  # closed file.
```

## 10. using `with` statement [Best of all]

`with` automatically closes the file.

syntax

```python
with open("C:\\Users\\User\\Desktop\\data.txt","r") as file:
    print(file.read())  
```

## 1. Reading PDF Files in Python

### What is a PDF?
* Portable Document Format.
* Used for reports, invoice, resumes
* Text is not directly readable --> needs a library


```python
import PyPDF2
```

```python
file = open(r"C:\Users\User\Downloads\file-sample_150kB.pdf","rb")  # open the file

reader = PyPDF2.PdfReader(file)  # binary format to readable format

print("Number of Pages:")
print(reader)  # file object
print("Number of Pages in the pdf:",len(reader.pages))

page = reader.pages[0]
print(page.extract_text())

print("===========page2===================")
page2 = reader.pages[1]
print(page2.extract_text())

print("=====================page3=================")
page3 = reader.pages[2]
print(page3.extract_text())

print("==================page4====================")
page4 = reader.pages[3]
print(page4.extract_text())
```

### Readling All Pages at once

----



```python
import PyPDF2

with open(r"C:\Users\User\Downloads\file-sample_150kB.pdf",'rb') as file:
    reader = PyPDF2.PdfReader(file)

    for page in reader.pages:
        print(page.extract_text())
```

```python
# total characters in this file
# total  Baba in this file 
file  = "NurseryRhymes.txt"

totalChar = 0
withSpace = 0
totalBaa = 0
with open(file,'r') as file:
    # print(file.read())
    content = file.read()

    words = content.split()
    # print(word)
    for ch in content:
        if ch == " ":
            withSpace+=1
        totalChar+=1
    for word in words:
        # print(word)
        if word  == "Baa," or word == "baa,":
            totalBaa+=1
        # print(ch)
print("Total Chracters in this file is:",totalChar)
print("Total Space in this file is:",withSpace)
print("Total Actual Characters in this file is:",totalChar-withSpace)
print("Total Baa in this file is:",totalBaa)
```

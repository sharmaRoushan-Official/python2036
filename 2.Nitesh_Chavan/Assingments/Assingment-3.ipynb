# 7. Write a function factorial(n) that returns factorial of a number.

```python
def factorial(n):
    fact = 1

    for i in range(1 , n+1):
        fact *= i

    return fact

factorial(5)
```

# 8. Write a function sum_list(lst) that returns the sum of all elements in a list.

```python
lst  = [1,2,3,4,5,6,7,8,9]

def sum_list(lst):
    sum = 0
    for i in lst:
        sum += i
    return sum

sum_list(lst)
```

# 9.Write a function count_vowels(s) that counts vowels in a string.

```python
str1 = "nitesh chavan"

def count_vowels(str1):
    vowels = "a","e","i","o","u"
    count = 0

    for ch in str1 :
        if ch in vowels :
            count += 1
    return count 

count_vowels(str1)
```

# 10. Write a function reverse_string(s) that returns the reverse of the string.


```python
str1 = "nitesh"

def reverse_string(str1):
    return str1[::-1]

reverse_string(str1)
```

# 11. Write a function is_palindrome(s) that checks whether the string is palindrome or not.

```python
str1 = "nitesh"

def is_palindrome(str1):
    str2 =""
    for ch in str1 :
        str2 = ch + str2

    if str1 == str2 :
        return "yes"
    else :
        return "no"

is_palindrome(str1)
```

## 24. Remove all negative numbers from a list using a loop.

```python
numbers = [10, -5, 3 , -1 , 0, 42, -88]

i=0

positiveList = []
negativeList = []


while i < len(numbers):
    if numbers[i] >= 0:
        positiveList.append(numbers[i])
    else :
        negativeList.append(numbers[i])
    i+=1
print("positive List = ", positiveList)
print("negative List = ", negativeList)
```

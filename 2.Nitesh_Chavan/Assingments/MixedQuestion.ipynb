## 1. Accept a string from the user and store each character in a list.


```python
name = input("Enter a string: ")

char_list = []

for ch in name:
    char_list.append(ch)

print("List of characters:", char_list)
```

## 2.Convert a list of names into a tuple.

```python
names = ["Nitesh", "Rahul", "Priya", "Amit"]

names_tuple = tuple(names)

print(names_tuple)
```

## 3. Find common elements between two lists.


```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

common = []

for item in list1:
    if item in list2:
        common.append(item)

print("Common elements:", common)
```

## 4.Convert a tuple into a string

```python
tup = ("H", "e", "l", "l", "o")

string = "".join(tup)

print("String:", string)
```

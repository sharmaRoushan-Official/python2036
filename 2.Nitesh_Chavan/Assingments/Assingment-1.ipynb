## 29. Write a Python program to classify *BMI (Body Mass Index)*:

    * <18.5 → Underweight
    * 18.5–24.9 → Normal
    * 25–29.9 → Overweight
    * ≥30 → Obese

```python
# Program to classify BMI (Body Mass Index)

weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

bmi = weight / (height ** 2)    #**2 means square of height 

print("BMI =", round(bmi, 2))  #round means rounded to 2 decial places

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal")
elif bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

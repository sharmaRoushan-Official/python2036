## 1.Given nums=[12,5,18,7,25,10] use map() to add each element with 10.

```python
nums = [12, 5, 18, 7, 25, 10]

def data(x):
    return x + 10

result = map(data, nums)

print(list(result))
```

```python
nums = [12, 5, 18, 7, 25, 10]
data = list(map(lambda x:x+10 ,nums))
print(data)
```

## 5. Given:
    nums = [5,8,2,10,3]
    Use reduce() with lambda to find the maximum numbers.


```python
from functools import reduce
```

```python
nums = [5,8,2,10,3]
def data(x,y):
    if x > y :
        return x
    else :
        return y
result = reduce(data, nums)
print(result)
```

```python
nums = [5,8,2,10,3]
data = reduce(lambda x,y: x if x>y else y , nums)
print(data)
```

## 7. Given:
    students =[
        ("Amit",85),
        ("Ravi",72),
        ("Neha",91),
        ("Pooja",78)
    ]
    Sort the list based on marks in ascending order.

```python
students = [
    ("Amit", 85),
    ("Ravi", 72),
    ("Neha", 91),
    ("Pooja", 78)
]

def marks(student):
    return student[1]

result = sorted(students, key=marks)

print(result)
```

```python
students = [
    ("Amit", 85),
    ("Ravi", 72),
    ("Neha", 91),
    ("Pooja", 78)
]

result = sorted( students, key= lambda x:x[1])
print(result)
```

## 9. Given:

        {"name":"A","salary":35000},
        {"name":"B","salary":50000},
        {"name":"C","salary":42000}

```python
employees = [
    {"name": "A", "salary": 35000},
    {"name": "B", "salary": 50000},
    {"name": "C", "salary": 42000}
]

def salary(x):
    return x["salary"]

result = sorted(employees, key=salary, reverse=True)

print(result)
```

### 19. JSON in Python

#### What is JSON?

JSON = JavaScript Object Notation

It is a lightweigh data format used to store and exchange data.

📌 Used in:
* Web APIs
* Dajngo/Flask
* Frontend ⮂ Backend communication
* Config Files


----
JSON vs Python Dictionary

JSON Looks like a Python dictioanry but there are differences

|Feature|JSON|Python Dictionary|
|----|----|----|
|`Format`|Text format|Python Object|
|`Quotes`|Keys must be in double quotes ""| Keys can be in single or double quotes|
|`Data Type`|Limited types| Supports many types|
|`Used for`|Data Transfer/storage|Programming logic|

Python Dictionary

````python
data = {'name':"Gayatri","age":65}
print(data)
print(type(data))
````

JSON String

`{"name":"Gayatri","age":65}`

### 5. Convery Python Dictionary to JSON

This is Called Serialization

json.dumps() ➼ Converts python object to JSON String

````python
import json
````

````python
data = {
    "name":"Gayatri",
    'age':105,
    'city':"NOIDA"
}
print(data,type(data))
````

````python
json_data = json.dumps(data)
print(json_data)
print(type(json_data))

# not accessible
````

Note:

So, dumps() return JSON string.

### 6. Convert JSON to Python Dictionary

This is called Deserialization

json.loads() ➼ Convert JSON string to Python Object 

````python
json_string = '{"name": "Gayatri", "age": 105, "city": "NOIDA"}'
print(json_string,type(json_string))
````

````python
python_data  = json.loads(json_string)
print(python_data,type(python_data))
````

````python
python_data['name'] = "Gayatri Gaware"
python_data
````

#### 7. Wrting JSON Data to a File

----

json.dump ➼ Write python object into JSON File



````python
data = {
    "id":101,
    "name":"Gayatri Gaware",
    "skill":["Python","Django","JavaScript","Database"]
}
print(data,type(data))
````

##### create a file to store information 

````python
with open("student.json",'w') as file:
    json.dump(data,file)
````

### 8. Reading JSON Data From a File

json.load() ➼ Read JSON File into Python Object


````python
with open("student.json","r") as f:
    data = json.load(f)
    print(data)
    print(type(data))
````

### 9. dumps() vs dump() and lodas() vs load()

----
|Function|Used For| Input| Output|
|----|----|----|----|
|json.dumps()|Python ➼ JSON string| Python Object | JSON string|
|json.dump()| Python ➼ JSON file|Pytyhon Object+File|Writes to file|
|json.loads()|JSON string ➼ Python| JSON string| Python Object|

### 10. Preety Print JSON [Indentation]

By default JSON is one-line. 


````python
data  = {"name":"Geeta","age":105, "city":"Dwarka"}
print(data)

print(json.dumps(data, indent=4))
````

````python
## 11. Sort JSON Keys

data = {"b":2,"a":1,"c":3}
print(data)

print(json.dumps(data,indent=4, sort_keys=False))
````

````python
data = {"b":2,"a":1,"c":3}
print(data)

print(json.dumps(data,indent=4, sort_keys=True))
````

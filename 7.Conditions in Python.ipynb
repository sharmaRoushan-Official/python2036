### 2. Syntax of if-statement

✅ syntax:

if condition:
    
    # code executes if condition becomes True.

```python
age = 19

if age>=18:
    print("You are elegible to vote")
```

KeyPoints:

* Condition must return True/False
* Indentation is Mandatory/Block True/False
* No parentheses is required(unlike C/Java)

```python
age = 16

if age>=18:
    print("Adult")
else:
    print("Minor")
```

keypoint:

* else runs when condition is False
* Only one else allowed per if
* else is optional

### 4.if-elif-else  Ladder if 

---

```python
marks = int(input("Enter your marks: "))

### 90-100 "A", 80-90 "B", 70-80: "C", 60-70 -"D", 50-60-"E" 40-50 -"Fail"

if marks>=90 and marks<=100:
    print("Grade:A")
elif marks>=80 and marks<90:
    print("Grade:B")
elif marks>=70 and marks<80:
    print("Grade:C")
elif marks>=60 and marks<70:
    print("Grade:D")
elif marks>=50 and marks<60:
    print("Grade:E")
else:
    print("Fail")
```

### 10. Using in or not-in 
### Membership operator - [in, not-in]


```python
fruits = ["apple","banana","mango","strawberry","grapes"]

if 'apple' in fruits:
    print("Available")
else:
    print("Not Available")
```

```python
fruits = ["apple","banana","mango","strawberry","grapes"]

fruitsName = input("Enter any fruits: ")
if fruitsName in fruits:
    print("Available")
else:
    print("Not Available")
```
